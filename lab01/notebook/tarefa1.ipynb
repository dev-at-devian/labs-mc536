Fonte 1 - GINA - Arquivos CSV 
* [Policies](https://extranet.who.int/nutrition/gina/sites/default/filesstore/csv/policy/gina_policies.csv)
* [Actions](https://extranet.who.int/nutrition/gina/sites/default/filesstore/csv/action/gina_actions.csv)
* [Mechanisms](https://extranet.who.int/nutrition/gina/sites/default/filesstore/csv/mechanisms/gina_mechanisms.csv)
* [Commitments](https://extranet.who.int/nutrition/gina/sites/default/filesstore/csv/commitments/gina_commitments.csv)

```python
import requests
from io import StringIO
import pandas as pd

# Obtendo os arquivos csv dos endereços e guardando como strings
gina_policies_csv = requests.get("https://extranet.who.int/nutrition/gina/sites/default/filesstore/csv/policy/gina_policies.csv").content.decode("utf-8")
gina_actions_csv = requests.get("https://extranet.who.int/nutrition/gina/sites/default/filesstore/csv/action/gina_actions.csv").content.decode("utf-8")
gina_mechanisms_csv = requests.get("https://extranet.who.int/nutrition/gina/sites/default/filesstore/csv/mechanisms/gina_mechanisms.csv").content.decode("utf-8")
gina_commitments_csv = requests.get("https://extranet.who.int/nutrition/gina/sites/default/filesstore/csv/commitments/gina_commitments.csv").content.decode("utf-8")

# Convertendo as strings em formato csv para um data frame, convertendo antes para um "arquivo" pelo StreamIO para serem usadas como input em read_csv
gina_policies = pd.read_csv(StringIO(gina_policies_csv))
gina_actions = pd.read_csv(StringIO(gina_actions_csv))
gina_mechanisms = pd.read_csv(StringIO(gina_mechanisms_csv))
gina_commitments = pd.read_csv(StringIO(gina_commitments_csv))

print(gina_policies.head) # Testando um dos dataframes
```

Fonte 2 - COVID-19 API - Arquivo JSON
* [URL](https://api.covid19api.com/summary)

```python
import requests
from io import StringIO
import json

# Obtendo o arquivo json do endereço e guardando como string
json_str = requests.get("https://api.covid19api.com/summary").content.decode("utf-8")

# Passando a string para a biblioteca interpretar e retornar um objeto json
js = json.loads(json_str)

print(js) # Testando o objeto
```
